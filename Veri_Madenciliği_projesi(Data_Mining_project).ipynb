```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C://Users//Hp//Desktop//DataMarket//Mall_Customers.csv")
```

1-Cinsiyete göre müşteri gruplarının yaş dağılımı ve ortalama yıllık gelirlerini analiz etmek?

```python
grouped=df.groupby("Gender").mean()
```

```python
plt.hist(df[df["Gender"]=="Male"]["Age"],alpha=0.5,label="Male")
plt.hist(df[df["Gender"]=="Female"]["Age"],alpha=0.5,label="Female")
plt.legend()
plt.xlabel("Age")
plt.ylabel("Number")
plt.title("Cinsiyete göre yaş dağılımı")
plt.show()
```

2-Müşterilerin yıllık gelir ve harcama skorları arasındaki ilişkiyi analiz etmek?


```python
import seaborn as sns

sns.scatterplot(x="Annual Income (k$)", y="Spending Score (1-100)", data=df)
plt.xlabel("Yıllık Gelir (bin $)")
plt.ylabel("Harcama Skoru (1-100)")
plt.title("Yıllık gelir ve harcama skoru arasındaki ilişki")
plt.show()

```

3-Kümeleme (Clustering) yöntemi kullanarak müşterileri benzer özelliklere sahip gruplara ayırmak?

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
kmeans.fit(X)

df['Cluster'] = kmeans.labels_

sns.scatterplot(x="Annual Income (k$)", y="Spending Score (1-100)", hue="Cluster", data=df, palette='rainbow')
plt.xlabel("Yıllık Gelir (bin $)")
plt.ylabel("Harcama Skoru (1-100)")
plt.title("Kümeleme sonucu müşteri grupları")
plt.show()
```

4-Cinsiyet ve yaşa göre müşterilerin harcama skorlarını analiz etmek?


```python
grouped = df.groupby(["Gender", "Age"]).mean()
grouped = df.groupby(['Gender', 'Age'], as_index=False)['Spending Score (1-100)'].mean()

sns.barplot(x=grouped.index, y='Spending Score (1-100)', data=grouped.reset_index())
plt.xlabel("Cinsiyet ve Yaş")
plt.ylabel("Ortalama Harcama Skoru (1-100)")
plt.title("Cinsiyet ve yaşa göre ortalama harcama skoru")
plt.xticks(rotation=90)
plt.show()
```

5-Veri setindeki ortalama yaşı bulmak?


```python
mean_age = df["Age"].mean()
print("Ortalama Yaş: ", mean_age)
```

6-Veri setindeki erkek ve kadınların sayısını bulmak?

```python
gender_counts = df["Gender"].value_counts()
print("Cinsiyet Sayıları: \n", gender_counts)
```

7-Veri setindeki en yüksek ve en düşük gelir değerlerini bulmak ?

```python
min_income = df["Annual Income (k$)"].min()
max_income = df["Annual Income (k$)"].max()
print("En Düşük Gelir: ", min_income, "k$")
print("En Yüksek Gelir: ", max_income, "k$")
```

8-Veri setindeki müşterilerin harcama puanlarını histogram ile görselleştirmek?


```python
spending_scores = df["Spending Score (1-100)"]

plt.hist(spending_scores, bins=10)
plt.xlabel("Harcama Puanı (1-100)")
plt.ylabel("Müşteri Sayısı")
plt.title("Harcama Puanı Dağılımı")
plt.show()
```

9-Veri setindeki müşterilerin yaşları ve harcama puanları arasındaki ilişkiyi görselleştirmek?

```python
plt.scatter(df["Age"], df["Spending Score (1-100)"])
plt.xlabel("Yaş")
plt.ylabel("Harcama Puanı (1-100)")
plt.title("Yaş ve Harcama Puanı İlişkisi")
plt.show()
```
